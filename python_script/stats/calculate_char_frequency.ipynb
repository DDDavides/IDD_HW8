```python
datasets_path = "../../csv_datasets"
datasets_cleaned_path = "../../csv_datasets_cleaned"

caratteri_bizzarri_path = "/caratteri_bizzarri.csv"
caratteri_bizzarri_per_team_path = "/caratteri_bizzarri_per_team.csv"

# pre cleaning
pre_dir_path = "./csv/pre_cleaning"
pre_bizzarre_path = pre_dir_path + caratteri_bizzarri_path # ./csv/pre_cleaning/caratteri_bizzarri.csv
pre_bizzarre_per_team_path = pre_dir_path + caratteri_bizzarri_per_team_path # ./csv/pre_cleaning/caratteri_bizzarri_per_team.csv

# post cleaning
post_dir_path = "./csv/post_cleaning"
post_bizzarre_path = post_dir_path + caratteri_bizzarri_path # ./csv/post_cleaning/caratteri_bizzarri.csv
post_bizzarre_per_team_path = post_dir_path + caratteri_bizzarri_per_team_path # ./csv/post_cleaning/caratteri_bizzarri_per_team.csv
```

```python
# Funzione che calcola la frequenza dei caratteri nei datasets presenti in una cartella
# N.B. i datasets devono essere in formato .csv
def calculate_char_freq(datasets_path):
    occurrences = {} 
    nameteam2occurrences = {} # dizionario di dizionario e.g. {"DDD": {<char1>: #1, <char2>: #2, ...}}
    for file in os.listdir(datasets_path):
        file_name = f"{datasets_path}/{file}"
        team_name = file.split("-")[0]

        if team_name not in nameteam2occurrences:
            nameteam2occurrences[team_name] = {}

        df = pd.read_csv(file_name)
        numColumns = len(df.columns)
        columnsName = df.columns
        for idx, _ in df.iterrows():
            for colIdx in range(numColumns):
                cellData = str(df.loc[idx, columnsName[colIdx]])
                for char in cellData:
                    if char not in nameteam2occurrences[team_name]:
                        nameteam2occurrences[team_name][char] = 0
                    if char not in occurrences:
                        occurrences[char] = 0
                    occurrences[char] += 1
                    nameteam2occurrences[team_name][char] += 1
    return occurrences, nameteam2occurrences
```

```python
allowed_char =  printable[:-38] + " .,"
# allowed_char = "\n\r"
```

# Calcolo occorrenze dei caratteri prima del cleaning

```python
before_occurrences, before_nameteam2occurrences = calculate_char_freq(datasets_path)
```

```python
display(pd.DataFrame(before_nameteam2occurrences))
```

```python
pre_bizzarre2occurrences = {key: [before_occurrences[key]] for key in before_occurrences if key not in allowed_char}
pre_bizzarre2occurrences = pd.DataFrame(pre_bizzarre2occurrences).T
display(pre_bizzarre2occurrences)
```

```python
pre_bizzare_per_team = { team : { key : before_occurrences[team][key] for key in before_occurrences[team] if key not in allowed_char} for team in before_occurrences}
pre_bizzare_per_team = pd.DataFrame(pre_bizzare_per_team)
```

```python
display(pre_bizzare_per_team)
```

```python
pre_bizzarre2occurrences.to_csv(pre_bizzarre_path, index=[0])
```

```python
pre_bizzare_per_team.to_csv(pre_bizzarre_per_team_path)
```
