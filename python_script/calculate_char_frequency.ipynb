```python
import pandas as pd
from string import printable
import os 
```

```python
mediated_schema_path = "./schema_matching/csv/schema_final.csv"
mediated_schema = pd.read_csv(mediated_schema_path)
```

```python
occurrences = {}
name2occurrences = {}
datasets_path = "../json_datasets"
for file in os.listdir(datasets_path):
    file_name = f"{datasets_path}/{file}"
    team_name = file.split("-")[0]

    if team_name not in name2occurrences:
        name2occurrences[team_name] = {}

    df = pd.read_json(file_name)
    for _, row in df.iterrows():
        for char in list(row.to_string()):
            
            if char not in name2occurrences[team_name]:
                name2occurrences[team_name][char] = 0
            if char not in occurrences:
                occurrences[char] = 0
            occurrences[char] += 1
            name2occurrences[team_name][char] += 1
```

```python
display(pd.DataFrame(name2occurrences))
```

```python
allawed_char =  printable[:-38] + " .,"
```

```python
bizzarre = {key.encode("utf-8"): [occurrences[key]] for key in occurrences if key not in allawed_char}
bizzarre = pd.DataFrame(bizzarre).T
display(bizzarre)
```

```python
bizzarre.to_csv("./caratteri_bizzarri.csv", index=[0])
```

```python
bizzare_per_team = { team : { key.encode("utf-8") : name2occurrences[team][key] for key in name2occurrences[team] if key not in allawed_char} for team in name2occurrences}
bizzare_per_team = pd.DataFrame(bizzare_per_team)
```

```python
bizzare_per_team.to_csv("./caratteribizzarri_per_team.csv")
```
